 # 1) What is the most costly college? What is the cheapest?

```python
dt['COSTT4_A']
```

```python
# combining program and annual cost together 
dt['COST'] = dt['COSTT4_A']
dt['COST'] = dt['COST'].fillna(dt['COSTT4_P'])
```

```python
print ('The most cost college is:',dt.loc[dt['COST'].idxmax()].INSTNM)
print ('For the cost of :',dt.loc[dt['COST'].idxmax()].COST)
```

```python
print ( 'The most cheap college is:',dt.loc[dt['COST'].idxmin()].INSTNM)
```

#  2) What is the average cost for college for colleges in different parts of the US?

```python
# Creating a data frame named AVgCost
# Dropping the row Named COST.
AvgCost = dt.dropna(subset=['COST'])
AvgCost = round(AvgCost.groupby(['STABBR'])['COST'].mean())
```

```python
print (AvgCost.sort_values(ascending=False))
```

 # 3) What is the average cost for college for religious vs. secular institutions?

```python
# Creating a new dataframe Recost for the cost of Religious instutions and droppig the rows with  missing value.
ReCost = dt.dropna(subset=['RELAFFIL'])
```

```python
print ( 'The average cost for religious college  is = ' ,round(ReCost.COST.mean()))
```

```python
# Creating a new dataframe SeCost for the cost of secular instutions
# dropping the column where secular is Nan.
seCost = dt.loc[pd.isnull(dt).any(1),:]
print ( 'The average cost for Secular college  is = ' , round(seCost.COST.mean()))
```
